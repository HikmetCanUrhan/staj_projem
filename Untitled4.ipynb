```python

# =============================== Dosya yolları ===============================

SEG_PATH = "/Users/hikmetcanurhan/Downloads/fake_station_data_100.xlsx"
VEH_PATH = "/Users/hikmetcanurhan/Downloads/fake_neighborhood_vehicle_data_300.xlsx"
OUT_PATH = "/Users/hikmetcanurhan/Desktop/github/stations_segmentasyon.xlsx"
SEG_SHEET = 0
VEH_SHEET = 0
# =============================================================================


# Türkçe -> küçük harf + trim + İ/I düzeltmesi + çoklu boşlukları tekleme
def normalize_basic(x):
    if pd.isna(x):
        return ""
    s = str(x).strip()
    s = s.replace("İ", "i").replace("I", "ı").lower()
    s = " ".join(s.split())
    return s

# Sonda gelen tür eklerini temizleme (yalnızca sondan!):
# "mah", "mah.", "mahalle", "mh", "mhl"
_mahalle_suffix_re = re.compile(r"\s+(mah(\.|alle)?|mh|mhl)$", flags=re.IGNORECASE)
def strip_mahalle_suffix(s):
    return _mahalle_suffix_re.sub("", s)

# Türkçe karakterleri ASCII'ye katlama (yüreğir -> yuregir)
_tr2ascii = str.maketrans({
    "ç":"c","ğ":"g","ı":"i","ö":"o","ş":"s","ü":"u",
    "â":"a","î":"i","û":"u","ê":"e","ô":"o",
})
def fold_ascii(s):
    return s.translate(_tr2ascii)

# Bazı il adları için tarihsel/alternatif ad düzeltmeleri (ihtiyaten)
PROVINCE_ALIASES = {
    "içel": "mersin",
    "urfa": "şanlıurfa",
    "afyon": "afyonkarahisar",
    "maraş": "kahramanmaraş",
    "antep": "gaziantep",
}

# Verileri okuma
seg = pd.read_excel(SEG_PATH, sheet_name=SEG_SHEET)  # istasyonlar
veh = pd.read_excel(VEH_PATH, sheet_name=VEH_SHEET)  # mahalle araç sayıları

# Kolonları ortak isimlere eşleme  (DÜZELTİLDİ)
seg = seg.rename(columns={"province": "il", "county": "ilçe", "neighbourhood_name": "mahalle"})
veh = veh.rename(columns={"province": "il", "county": "ilçe", "neighbourhood_name": "mahalle"})

#  Anahtar kolonları normalize etme
for df_ in (seg, veh):
    for col in ["il", "ilçe", "mahalle"]:
        df_[col] = df_[col].map(normalize_basic)

# İl alias düzeltmeleri
for df_ in (seg, veh):
    df_["il"] = df_["il"].map(lambda x: PROVINCE_ALIASES.get(x, x))

# Mahalle son eklerini (yalnızca sondan) kırpma
for df_ in (seg, veh):
    df_["mahalle_clean"] = df_["mahalle"].map(strip_mahalle_suffix)

# Birincil (TR) ve ikincil (ASCII-katlanmış) anahtarlar
for df_ in (seg, veh):
    df_["il_k1"]      = df_["il"]
    df_["ilçe_k1"]    = df_["ilçe"]
    df_["mahalle_k1"] = df_["mahalle_clean"]

    df_["il_k2"]      = df_["il_k1"].map(fold_ascii)
    df_["ilçe_k2"]    = df_["ilçe_k1"].map(fold_ascii)
    df_["mahalle_k2"] = df_["mahalle_k1"].map(fold_ascii)

KEYS1 = ["il_k1", "ilçe_k1", "mahalle_k1"]
KEYS2 = ["il_k2", "ilçe_k2", "mahalle_k2"]

# MAHALLE datasını tekilleştirme  (DÜZELTİLDİ: final_score yerine total_vehicle_count)
veh_reduced_k1 = (
    veh.groupby(KEYS1, as_index=False, dropna=False)
       .agg(total_vehicle_count=("total_vehicle_count","mean"))
)
veh_reduced_k2 = (
    veh.groupby(KEYS2, as_index=False, dropna=False)
       .agg(total_vehicle_count_k2=("total_vehicle_count","mean"))
)

# Kademeli merge: önce K1, sonra K2 ile tamamlama
merged = seg.merge(veh_reduced_k1, on=KEYS1, how="left")
merged = merged.merge(veh_reduced_k2, on=KEYS2, how="left")

# Nihai metrik: mahalleden gelen toplam araç sayısı
merged["total_vehicle_count"] = merged["total_vehicle_count"].combine_first(merged["total_vehicle_count_k2"])

# Yardımcı kolonları temizleme
drop_cols = ["mahalle_clean","il_k1","ilçe_k1","mahalle_k1","il_k2","ilçe_k2","mahalle_k2","total_vehicle_count_k2"]
merged = merged.drop(columns=[c for c in drop_cols if c in merged.columns])

# Çıktı
merged.to_excel(OUT_PATH, index=False)

# Rapor
total = len(seg)
matched = merged["total_vehicle_count"].notna().sum()
unmatched = total - matched
print(f"Tamamlandı. Çıktı: {OUT_PATH}")
print(f"Eşleşen satır: {matched} / {total}  (oran: {matched/total:.2%})")
print(f"Eşleşmeyen satır: {unmatched}")

# Eşleşmeyenlerden örnek anahtarlar
if unmatched > 0:
    sample_missing = (
        merged.loc[merged["total_vehicle_count"].isna(), ["il","ilçe","mahalle"]]
              .drop_duplicates()
              .head(20)
    )

# TÜM eşleşmeyenleri yazdır
unmatched_full = merged.loc[merged["total_vehicle_count"].isna(), ["il", "ilçe", "mahalle"]]
print(f"\nToplam eşleşmeyen satır sayısı: {len(unmatched_full)}")
with pd.option_context("display.max_rows", None, "display.max_colwidth", None):
    print(unmatched_full.to_string(index=False))
```

```python
# ==== Alt ve üst sınırlar (manuel belirlenen değerler) =====
thresholds = {
    "daily_vehicle_count": (100, 1000),               # min, max
    "daily_motor_count": (0, 250),
    "total_car_count": (500, 30000),
    "avg_car_count": (100, 500),
    "total_vehicle_count" : (700, 3250)
}

# ==== Outlier indexlerini bulma ====
outlier_idx = set()
for f, (min_val, max_val) in thresholds.items():
    idx = df[(df[f] < min_val) | (df[f] > max_val)].index
    outlier_idx.update(idx)
    print(f"{f} için outlier sayısı (<{min_val} veya >{max_val}):", len(idx))

print("Toplam outlier sayısı:", len(outlier_idx))

# ==== Outlier’sız temiz veri ====
df_no_outliers = df.drop(outlier_idx).reset_index(drop=True)
print("Temizlenmiş veri boyutu:", df_no_outliers.shape)

# ==== Temizlenmiş grafikleri çizme ====
for f in thresholds.keys():
    fig, axes = plt.subplots(1, 2, figsize=(10,4), sharey=True)
    
    # ==== Orijinal veri ====
    sns.boxplot(y=df[f], ax=axes[0], color="lightcoral")
    axes[0].set_title(f"{f} (Orijinal)")
    
    # ==== Outlier’sız veri ====
    sns.boxplot(y=df_no_outliers[f], ax=axes[1], color="lightgreen")
    axes[1].set_title(f"{f} (Temizlenmiş)")
    
    plt.suptitle(f"{f} için Outlier Öncesi vs Sonrası", fontsize=12)
    plt.show()
```

```python
# ==== Feature'lar seçdi ====
X = df_no_outliers[features]   # outlier'sız veri

# ==== StandardScaler ile ölçekleme ====
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Ölçeklenmiş verinin boyutu:", X_scaled.shape)

# Kontrol için ilk 5 satır (numpy array olarak)
print(X_scaled[:5])
```

```python
# =================================================================
# Elbow (Inertia) Analizi
# Amaç: k'ya göre SSE (inertia) nasıl düşüyor, "dirsek" nerede?
# Girdi: X_scaled (StandardScaler sonrası, outliersız)
# =================================================================


# ==== k aralığını belirleme ====
K_RANGE = range(2, 20)  # 2..10 arası dene (istersen değiştir)

# ==== Her k için KMeans fit et ve inertia (SSE) toplama ====
inertias = []
for k in K_RANGE:
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    km.fit(X_scaled)
    inertias.append(km.inertia_)

# ==== Grafiği çizme ====
plt.figure(figsize=(7,4.5))
plt.plot(list(K_RANGE), inertias, marker="o")
plt.title("Elbow (Inertia) Grafiği")
plt.xlabel("Küme Sayısı (k)")
plt.ylabel("Inertia (Toplam Kare Hata, SSE)")
plt.grid(alpha=0.3)
plt.xticks(list(K_RANGE))
plt.show()


```

```python
# ============================================================
# Silhouette Analizi
# Amaç: her k için ortalama Silhouette skoru ölçmek
# Girdi: X_scaled (outlier'sız + StandardScaler sonrası)
# ===========================================================

sil_scores = []
K_RANGE = range(2, 11)  # 2'den 10 kümeye kadar dene

for k in K_RANGE:
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    labels = km.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    sil_scores.append(score)

# Grafiği çizme
plt.figure(figsize=(7,4.5))
plt.plot(list(K_RANGE), sil_scores, marker="o", color="purple")
plt.title("Silhouette Skorları (k karşılaştırması)")
plt.xlabel("Küme Sayısı (k)")
plt.ylabel("Ortalama Silhouette Skoru")
plt.grid(alpha=0.3)
plt.xticks(list(K_RANGE))
plt.show()

# En iyi k'yi yazdırma
best_k = list(K_RANGE)[int(np.argmax(sil_scores))]
print("En iyi Silhouette skoruna göre k =", best_k, 
      " | skor =", max(sil_scores).round(3))

```

```python
# ================== PLOTLY 3D ==================

# ==== PCA -> 3B projeksiyon ve merkezler ====
pca = PCA(n_components=3, random_state=42)
X_3d = pca.fit_transform(X_scaled)
centers_3d = pca.transform(kmeans.cluster_centers_)
var1, var2, var3 = pca.explained_variance_ratio_

# ==== Veri çerçeveleri ====
plot_df = pd.DataFrame(X_3d, columns=["Dim1","Dim2","Dim3"])
plot_df["cluster"] = df_no_outliers["segment"].values
centers_df = pd.DataFrame(centers_3d, columns=["Dim1","Dim2","Dim3"])

segments = sorted(plot_df["cluster"].unique())

# ==== Renk paleti ====
palette = {
    0:"#1f77b4", 1:"#2ca02c", 2:"#9467bd", 3:"#e377c2",
    4:"#17becf", 5:"#bcbd22", 6:"#8c564b", 7:"#ff7f0e"
}

# ==== Nokta izleri (cluster bazında) ====
traces = []
for seg in segments:
    pts = plot_df.loc[plot_df["cluster"]==seg, ["Dim1","Dim2","Dim3"]].to_numpy()
    c = palette.get(seg, "#999999")
    traces.append(
        go.Scatter3d(
            x=pts[:,0], y=pts[:,1], z=pts[:,2],
            mode="markers",
            marker=dict(size=3, opacity=0.85, color=c),
            name=f"Küme {seg}",
            legendgroup=f"g{seg}",
            showlegend=True
        )
    )

# ==== Merkezler ====
traces.append(
    go.Scatter3d(
        x=centers_df["Dim1"], y=centers_df["Dim2"], z=centers_df["Dim3"],
        mode="markers+text",
        marker=dict(size=6, color="yellow", line=dict(width=2, color="black")),
        text=[f"C{i}" for i in range(len(centers_df))],
        textposition="top center",
        name="Merkezler",
        showlegend=True
    )
)

# ==== Convex Hull (Mesh3d) — başlangıçta GİZLİ ====
hull_traces = []
for seg in segments:
    pts = plot_df.loc[plot_df["cluster"]==seg, ["Dim1","Dim2","Dim3"]].to_numpy()
    if len(pts) >= 4:
        try:
            hull = ConvexHull(pts)
            i, j, k = hull.simplices.T
            c = palette.get(seg, "#999999")
            hull_traces.append(
                go.Mesh3d(
                    x=pts[:,0], y=pts[:,1], z=pts[:,2],
                    i=i, j=j, k=k,
                    color=c, opacity=0.15,
                    name=f"Hull {seg}",
                    legendgroup=f"g{seg}",
                    showlegend=False,
                    visible=False
                )
            )
        except Exception:
            pass

# ==== Şekli oluşturma ====
fig = go.Figure(data=traces + hull_traces)

# ==== Yerleşim: legend altta yatay, butonlar sol üstte ====
num_scatter = len(traces)
num_hulls   = len(hull_traces)

fig.update_layout(
    title=dict(
        text="Küme Görselleştirme 3D (Plotly) — Hull Toggle",
        x=0.5,                 # başlığı ortala
        xanchor="center"
    ),
    scene=dict(
        xaxis_title=f"Dim1 ({var1*100:.1f}%)",
        yaxis_title=f"Dim2 ({var2*100:.1f}%)",
        zaxis_title=f"Dim3 ({var3*100:.1f}%)",
        aspectmode="data"
    ),
    legend=dict(
        orientation="h",
        x=0, y=-0.12, xanchor="left", yanchor="top",
        bgcolor="rgba(255,255,255,0.8)"
    ),
    margin=dict(l=0, r=0, t=60, b=60),

    width =1000,
    height = 850,

    updatemenus=[dict(
        type="buttons",
        direction="right",  
        x=0.02, y=1.10, xanchor="left", yanchor="top",
        buttons=[
            dict(
                label="Hull Aç",
                method="update",
                args=[{"visible": [True]*num_scatter + [True]*num_hulls}]
            ),
            dict(
                label="Hull Kapat",
                method="update",
                args=[{"visible": [True]*num_scatter + [False]*num_hulls}]
            ),
        ],
        bgcolor="white",
        bordercolor="#d0d0d0",
        borderwidth=1,
        pad={"r":6, "t":4, "b":4, "l":6}
    )]
)

fig.show()
```
